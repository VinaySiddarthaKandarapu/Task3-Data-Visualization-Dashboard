```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Load
df = pd.read_csv(r'C:\Users\vinay\OneDrive\Desktop\Intership\data\raw\supermarket_sales.csv')

# Clean
df.drop_duplicates(inplace=True)
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
df['date'] = pd.to_datetime(df['date'])

print("✅ Done!")
print("Shape:", df.shape)
print("Columns:", df.columns.tolist())
```

```python
# Check column names
print(df.columns.tolist())
```

```python
# Sales trend by date
df_daily = df.groupby('date')['revenue'].sum().reset_index()

plt.figure(figsize=(12,5))
plt.plot(df_daily['date'], df_daily['revenue'], color='steelblue')
plt.title('Daily Sales Trend')
plt.xlabel('Date')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig(r'C:\Users\vinay\OneDrive\Desktop\Intership\reports\sales_trend.png')
plt.show()
print("✅ Saved!")
```

```python
plt.figure(figsize=(10,5))
df.groupby('product_line')['revenue'].sum().sort_values().plot(kind='barh', color='teal')
plt.title('Revenue by Product Line')
plt.xlabel('Total Revenue')
plt.tight_layout()
plt.savefig(r'C:\Users\vinay\OneDrive\Desktop\Intership\reports\revenue_by_category.png')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['rating'], bins=15, color='coral', edgecolor='black')
plt.title('Customer Rating Distribution')
plt.xlabel('Rating'); plt.ylabel('Frequency')
plt.savefig(r'C:\Users\vinay\OneDrive\Desktop\Intership\reports\rating_hist.png')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df['unit_cost'], df['quantity'], alpha=0.5, c='purple')
plt.title('Unit Cost vs Quantity Sold')
plt.xlabel('Unit Cost'); plt.ylabel('Quantity')
plt.savefig(r'C:\Users\vinay\OneDrive\Desktop\Intership\reports\price_vs_qty.png')
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14,10))
df.groupby('city')['revenue'].sum().plot(kind='bar', ax=axes[0,0], color='steelblue', title='Revenue by City')
axes[0,1].hist(df['rating'], bins=15, color='coral')
axes[0,1].set_title('Rating Distribution')
df.groupby('payment_method')['revenue'].sum().plot(kind='pie', ax=axes[1,0], autopct='%1.1f%%', title='Payment Method Share')
axes[1,1].scatter(df['unit_cost'], df['quantity'], alpha=0.5)
axes[1,1].set_title('Cost vs Quantity')
plt.tight_layout()
plt.savefig(r'C:\Users\vinay\OneDrive\Desktop\Intership\reports\subplots_overview.png')
plt.show()
```

```python
plt.figure(figsize=(10,7))
numeric_df = df.select_dtypes(include='number')
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.savefig(r'C:\Users\vinay\OneDrive\Desktop\Intership\reports\heatmap.png')
plt.show()
```
